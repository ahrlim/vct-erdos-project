In this notebook, I will create dataframes: one with kills and economy information per rounds and the other is player rating information.\
Goal is to define our own economy rating with kills and economy information per rounds and then use this economy rating together with player rating

```python
import pandas as pd
```

```python
rounds_kills = pd.read_csv("../data/vct_2023/matches/rounds_kills.csv")
eco_rounds = pd.read_csv("../data/vct_2023/matches/eco_rounds.csv")
rounds_kills["Round Number"] = rounds_kills["Round Number"].apply(lambda x: int(x.split(" ")[-1]))
eco_rounds["Outcome"] = eco_rounds["Outcome"].apply(lambda x: 1 if x == "Win" else -1)
eco_rounds["Loadout Value"] = eco_rounds["Loadout Value"].apply(lambda x: int(x.replace(".","").replace("k","00")))
eco_rounds["Remaining Credits"] = eco_rounds["Remaining Credits"].apply(lambda x: int(x.replace(".","").replace("k","00")))

maps_scores = pd.read_csv("../data/vct_2023/matches/maps_scores.csv")
overview = pd.read_csv("../data/vct_2023/matches/overview.csv")
players_stats = pd.read_csv("../data/vct_2023/players_stats/players_stats.csv")
```

```python
# The next two print values show that "rounds_kills" and "eco_rounds" don't have NaN.
print(rounds_kills.isna().sum().sum())
print(eco_rounds.isna().sum().sum())
```
